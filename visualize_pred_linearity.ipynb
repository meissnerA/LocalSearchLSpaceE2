```python
import os
import numpy as np
import pandas as pd
from numpy import genfromtxt
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.interpolate import UnivariateSpline
```

```python
sns.set_style('darkgrid') # darkgrid, white grid, dark, white and ticks

plt.rc('axes', titlesize=18)     # fontsize of the axes title
plt.rc('axes', labelsize=14)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=13)    # fontsize of the tick labels
plt.rc('ytick', labelsize=13)    # fontsize of the tick labels
plt.rc('legend', fontsize=13)    # legend fontsize
plt.rc('font', size=13)          # controls default text sizes
#sns.color_palette('deep')
```

```python
my_data = genfromtxt("./figure_5_6/seed_3.txt", delimiter=';')
my_data[:,0] = -my_data[:,0]
#my_data[:,1] = -my_data[:,1]
df = pd.DataFrame(my_data, columns=['l_vec_coeff', 'regressor_logit', 'attr_pres', 'ident_pres'])
plt.figure(figsize=(10,6), tight_layout=True)
ax = sns.scatterplot(data=df, x='l_vec_coeff', y='regressor_logit', palette='Set2', s=60)
ax.set(xlabel='l_vec_coeff', ylabel='regressor_logit')
plt.show()
#plt.savefig('logit_over_l_vec_coeff.png')
```

```python
pred_list = []
path = "./figure_5_6/"
for file in os.listdir(path)[0:1000]:
    my_data = genfromtxt(path+file, delimiter=';')
    pred_list.append(my_data)
```

```python
my_data = genfromtxt("./figure_5_6/seed_3.txt", delimiter=';')

my_data[:,0] = -my_data[:,0]
mydata = np.flipud(my_data)
df = pd.DataFrame(my_data, columns=['l_vec_coeff', 'regressor_logit', 'attr_pres', 'ident_pres'])
#plt.figure(figsize=(10,6), tight_layout=True)
fig, ax = plt.subplots(figsize=(10,6), tight_layout=True)

palette = {1:'red', 2:'black', 3:'green'}
df2 = pd.DataFrame(np.transpose(np.vstack([-np.array(pred_list)[:,:,0].mean(axis=0), 
                                           np.array(pred_list)[:,:,1].mean(axis=0)])), columns=['x', 'y'])
sns.scatterplot(data=df2, x='x', y='y', s=60, label="mean(seed=0..999)", palette=palette)
sns.scatterplot(data=df, x='l_vec_coeff', y='regressor_logit', s=60, label="seed=3")
ax.set(xlabel='l_vec_coeff', ylabel='regressor_logit')

ax.legend() 

#plt.show()
plt.savefig('logit_over_l_vec_coeff.png')
```

```python
my_data = genfromtxt("./figure_5_6/seed_3.txt", delimiter=';')

df = pd.DataFrame(my_data, columns=['l_vec_coeff', 'regressor_logit', 'attr_pres', 'ident_pres'])
fig, ax = plt.subplots(figsize=(10,6), tight_layout=True)


palette = {1:'red', 2:'black', 3:'green'}
df2 = pd.DataFrame(np.transpose(np.vstack([-np.array(pred_list)[:,:,0].mean(axis=0), 
                                           np.array(pred_list)[:,:,2].mean(axis=0)])), columns=['x', 'y'])
sns.scatterplot(data=df2, x='x', y='y', s=60, palette=palette)

ax.set(xlabel='l_vec_coeff', ylabel='attribute preservation')

ax.legend() 


#plt.show()
plt.savefig('attr_pres_over_l_vec_coeff.png')
```
